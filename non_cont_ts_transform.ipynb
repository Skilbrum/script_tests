```python
import pandas as pd
```

```python
idx = pd.date_range('2020-01-01 01', '2020-01-01 05', freq='10min')
gap = idx[idx.get_loc('2020-01-01 03')]
idx = idx.drop(gap)
s = pd.Series(range(idx.shape[0]), idx)
```

```python
s.rolling(3).mean()
```

```python
s.rolling('30min').mean()
```
